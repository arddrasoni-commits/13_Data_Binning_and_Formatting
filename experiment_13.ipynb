```python
import pandas as pd
import numpy as np
data = {
    'Product': ['tuner keys', 'Allen Key', 'Bridge nut', 'scale bar'],
    'Price' :[530, 390, 780, 1000],
    'Units_Sold' : [7,45,22,89]
}

df = pd.DataFrame(data)
print("Data Set")
df
```

```python
# Data Binning
bins = [0,200,500,1000]
labels = ['Low','Medium','High']

df['Price Category'] = pd.cut(df['Price'],bins = bins, labels=labels)
df
```

```python
# Binning Units Sold
bins2 = [0,30,60,100]
labels2 = ['Low Sales','Medium Sales','High Sales']

df ['Sales_Category']= pd.cut(df['Units_Sold'], bins = bins2, labels = labels2)
print(df)
```

```python
# Data Formatting
# Convert Data type

print(df.dtypes)

df['Units_Sold'] = df['Units_Sold'].astype(float)

print(df.dtypes)
```

```python
# Convert Text to Uppercase
df['Product'] = df['Product'].str.upper()

print(df)
```

```python
# Rounding Values
df['Price'] = df['Price'].round(2)
df
```

```python
# Additional operations
# Sort Data
sorted_df = df.sort_values(by = 'Price')
print(sorted_df)
```

```python
# Display Unique Values
print(df['Price Category'].unique())
```
